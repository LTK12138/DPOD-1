```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("big_models_grid.csv")
data2 = pd.read_csv("big_models_grid2.csv")
big_models = pd.read_csv("small_models_grid.csv")
```

```python
small_models = pd.concat([data1, data2])
small_models = small_models[small_models['mean_error'] < 10000]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(small_models['reprError'], small_models['accuracy'], label="Accuracy on low quality 3D model (less faces)")
# plt.plot(small_models['reprError'], small_models['mean_error'], label="Mean euclidean distance between estimated position of verice in 3D model and groung truth position; low qulaity 3D model")
plt.plot(big_models['reprError'], big_models['accuracy'])
plt.plot(big_models['reprError'], big_models['mean_error'])
plt.legend(loc="upper left")
```
